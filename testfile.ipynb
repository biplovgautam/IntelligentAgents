```python
# general Agent Workflow

current_temp = 20
goal_temp = 40

#function 1st
def percept():
    return ( current_temp)

# function 2nd
def process():
    if  current_temp <  goal_temp:
        return "higher"
    else:
        return "stop"
# function 3rd
def act(action):
    global current_temp
    if  action == "higher":
         current_temp += 5
    else:
        return None

while True:
    sense = percept()
    action = process()
    act(action)
    if action == "stop":
        break
    import time 
    time.sleep(0.5)
    print("current:",current_temp)
print("goal reached")
    
```
